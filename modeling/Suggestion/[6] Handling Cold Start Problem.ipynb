```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('../../data/datasets/movie/ml_detail.csv', sep=',', encoding='latin-1')
movies.info()
```

```python
def convert_2_num(x):
    x = str(x)
    x = x.upper()
    suffix = x[-1]
    if suffix == 'K':
        return float(x[:-1]) * 1000
    elif suffix == 'M':
        return float(x[:-1]) * 1000000
    elif suffix == 'B':
        return float(x[:-1]) * 1000000000
```

```python
movies['rating total'] = movies['rating total'].apply(convert_2_num)
```

```python
reliable_movies = movies[(movies['rating total'] >= 100000) & (movies['average rating'] >= 9)]
reliable_movies
```

```python
cold_start_suggest = pd.DataFrame(columns=['movie id'])
cold_start_suggest['movie id'] = ["|".join(reliable_movies['movie id'].tolist())]
```

```python
cold_start_suggest
```

```python
cold_start_suggest.to_csv('../../result/suggest_cold_start.csv', sep=',', encoding='latin-1', index=False)
```
