```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('../data/datasets/movie/details.csv', sep=',', encoding='latin-1')
movies.head()
```

```python
movies.info()
```

```python
movies = movies.dropna(subset=['movie id'])
movies = movies.dropna(subset=['title'])
```

```python
duplicated_movies = movies[movies.duplicated()]
duplicated_movies
```

```python
movies = movies.drop_duplicates(keep='first')
```

```python
duplicated_movies = movies[movies.duplicated()]
duplicated_movies
```

```python
movies.to_csv('../data/datasets/movie/cleaned_1_detail.csv', sep=',', encoding='latin-1', index=False)
```
