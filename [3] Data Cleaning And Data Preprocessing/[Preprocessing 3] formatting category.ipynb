```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('../data/datasets/movie/ml_detail.csv', sep=',', encoding='latin-1')
```

```python
movies.info()
```

```python
genres = set()
for i in movies['genre list']:
    genre_group = i.split("|")
    for genre in genre_group:
        genres.add(genre)
```

```python
genres
```

```python
genres = list(genres)
for genre in genres:
    movies[genre] = np.zeros(len(movies))
```

```python
for i in range(len(movies)):
    genre_group = movies.iloc[i]['genre list']
    for genre in genre_group.split("|"):
        movies.at[i, genre] = 1
```

```python
movies.head()
```

```python
movies.to_csv('../data/datasets/movie/ml_detail.csv', sep=',', encoding='latin-1', index=False)
```
