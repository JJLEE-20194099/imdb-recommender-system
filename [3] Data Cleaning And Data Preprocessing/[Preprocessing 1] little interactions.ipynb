```python
ratings = pd.read_csv('../data/datasets/rating/cleaned_1_detail.csv', sep=',', encoding='latin-1')
```

```python
ratings
```

```python
print(len(ratings[ratings['rating'] == 1]))
print(len(ratings[ratings['rating'] == 2]))
print(len(ratings[ratings['rating'] == 3]))
print(len(ratings[ratings['rating'] == 4]))
print(len(ratings[ratings['rating'] == 5]))
print(len(ratings[ratings['rating'] == 6]))
print(len(ratings[ratings['rating'] == 7]))
print(len(ratings[ratings['rating'] == 8]))
print(len(ratings[ratings['rating'] == 9]))
print(len(ratings[ratings['rating'] == 10]))
```

```python
import math
def normalizing_rating_score(score):
    return math.ceil(score / 2) 
```

```python
ratings['rating'] = ratings['rating'].apply(normalizing_rating_score)
```

```python
print(len(ratings[ratings['rating'] == 1]))
print(len(ratings[ratings['rating'] == 2]))
print(len(ratings[ratings['rating'] == 3]))
print(len(ratings[ratings['rating'] == 4]))
print(len(ratings[ratings['rating'] == 5]))
```

```python
no_ratings_per_user = ratings.groupby('user id').count()['rating']
```

```python
no_ratings_per_user
```

```python
user_ids = ratings.groupby('user id').count()['rating'].index.tolist()
user_ids = np.array(user_ids)
```

```python
user_ids
```

```python
no_ratings_per_user = np.array(no_ratings_per_user)
user_id_idxes = np.argsort(no_ratings_per_user).tolist()
```

```python
cnt = 0
for i in range(len(user_id_idxes[:1000])):
    idx = user_id_idxes[len(user_id_idxes) - i - 1]
    print(user_ids[idx], no_ratings_per_user[idx])
```

```python
eliminated_user_ids = user_ids[user_id_idxes[:-1000]].tolist() + user_ids[user_id_idxes[-10:]].tolist()
```

```python
eliminated_user_ids
```

```python
eliminated_user_ids_df = pd.DataFrame(columns=['user id'])
eliminated_user_ids_df['user id'] = eliminated_user_ids
```

```python
eliminated_user_ids_df.head()
```

```python
eliminated_ratings_df = pd.merge(ratings, eliminated_user_ids_df, how='inner', left_on='user id', right_on='user id')
eliminated_ratings_df
```

```python
eliminated_ratings_df.to_csv('../data/datasets/rating/dropped_detail.csv', sep=',', encoding='latin-1', index=False)
```

```python
remaining_user_ids = user_ids[user_id_idxes[-1400:-10]]
print(len(remaining_user_ids))
remaining_user_ids
```

```python
remaining_user_ids_df = pd.DataFrame(columns=['user id'])
remaining_user_ids_df['user id'] = remaining_user_ids
```

```python
remaining_user_ids_df.head()
```

```python
remaining_ratings_df = pd.merge(ratings, remaining_user_ids_df, how='inner', left_on='user id', right_on='user id')
remaining_ratings_df
```

```python
remaining_ratings_df['movie id'].unique()
```

```python
len(remaining_ratings_df['movie id'].unique())
```

```python
movie_ids = remaining_ratings_df['movie id'].value_counts().index.tolist()
```

```python
movie_ids
```

```python
len(movie_ids)
```

```python
movie_ids_df = pd.DataFrame(columns=['movie id'])
movie_ids_df['movie id'] = movie_ids
movie_ids_df
```

```python
merged_movie_ids_and_ratings = pd.merge(movie_ids_df, remaining_ratings_df, how='inner', left_on='movie id', right_on='movie id')
merged_movie_ids_and_ratings
```
